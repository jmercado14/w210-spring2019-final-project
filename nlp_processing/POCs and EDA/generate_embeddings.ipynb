```python
def preprocess_chat(text):
    text = text.lower()
    text = re.sub("\’\’|\'\'|\`\`", '"', text) # quotation marks (?)
    text = re.sub('’|`', "'", text) # apostrophes
    text = re.sub(':\)|:-\)|:\'\)|:\'-\)|:\]|:-\]|:d|:-d|=\)|=\]|xd|x-d|=d|8d|8\)|:3', 'emoticon_happy ', text) # emoticons
    text = re.sub(':\(|:-\(|:\'\(|:\'-\(|:\[|:-\[|d:|d-:|=\(|=\[|:c|:-c|:<|:-<', 'emoticon_sad', text) # emoticons
    text = re.sub(':p|:-p|=p|=-p', 'emoticon_playful', text) # emoticons
    text = re.sub(';\)|;-\)|;^\)|;\]|;-\]|;d|;-d]', 'emoticon_winking', text) # emoticons
    text = re.sub(':o|:-o|:0|:-0|=o|=0]', 'emoticon_surprised ', text) # emoticons
    text = re.sub(':\\|:-\\|:/|:-/|=\\|=/]', 'emoticon_meh ', text) # emoticons
    text = re.sub('[0-9]+', '', text) # digits
    text = re.sub('[!@#$\-\*/]+', ' ', text) # punctuation
    return text
```

```python
chats = chat_file.text

# preprocess chats
# THIS WILL TAKE TIME
preprocessed = []
for i in range(len(chats)):
# for i in range(500): # for testing
    if i%100000 == 0:
        print (i, "chats preprocessed:", datetime.datetime.now())
    #text = gensim.utils.simple_preprocess(chats[i])
    text = preprocess_chat(str(chats[i]))
    text = word_tokenize(text)
    preprocessed.append(text)
    
print ("done:", datetime.datetime.now())
```

```python
# this will take SEVERAL HOURS
model = Word2Vec(preprocessed, iter=3, size=50, workers=4) # num dimensions = 50
```
